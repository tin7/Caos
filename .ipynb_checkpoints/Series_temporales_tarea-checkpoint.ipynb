```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Cargo archivo de datos

```python
datos=pd.read_csv("/home/tincho/Desktop/Caos/Tp/TP3/osc15-EST-01.csv",sep=";",decimal=",")
datos.head()
```

Convierto en formato numpy

```python
datosNP=datos.to_numpy()
```

Lo guardo en vectores representativos

```python
Tiempo=np.float64(datosNP[:,0])
Patron=np.float64(datosNP[:,1])
R=np.float64(datosNP[:,2])
T=np.float64(datosNP[:,3])
```

Grafico para ver que onda?

```python
plt.figure(figsize=(15,8))
plt.plot(Tiempo,R)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(Tiempo,T)
plt.grid()
plt.show()
```

Calculo la matriz de covarianza

```python
X=np.c_[R,T]
```

```python
xraya=np.mean(X,0)
B=X-xraya
Bt=np.transpose(B)
N=len(R)
A=(Bt@B)/N
A
```

Evalúo los autovalores y autovectores

```python
E=np.linalg.eigh(A)
print(E[0])
print(E[1])
```

Coloco los autovectores como columnas

```python
V=np.flip(E[1],1)
```

Proyecto en las componentes principales y grafico

```python
Z=B@V
```

```python
z1=Z[:,0]
z2=Z[:,1]
```

```python
plt.figure(figsize=(15,8))
plt.plot(Tiempo,z1)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(Tiempo,z2)
plt.grid()
plt.show()
```

z2 tiene el ruido y z1 tiene la señal mas limpia

Grafico z1 versus el patrón para ver si me dice algo la figura de Lissajous


```python
plt.figure(figsize=(15,8))
plt.plot(Patron,z1)
plt.grid()
plt.show()
```

Indica la presencia de un sistema dinámico subyacente,,, 
pero para ser concluyente se debe mejorar la relación señal ruido


Para ello le hago un flitrado de datos

```python
from scipy.signal import butter, lfilter, freqz
def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a
def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y
# Parámetros de filtro.
order = 2
fs = 16377.87       # taza de muestreo, en Hz se lo sabe del equipo
cutoff = 1000.0  # frecuencia de corte, Hz
# coeficientes del filtro}
b, a = butter_lowpass(cutoff, fs, order)
```

```python
# ahora se filtra los datos,,,
z1_f = butter_lowpass_filter(z1, cutoff, fs, order)
```

```python
plt.figure(figsize=(15,8))
plt.plot(Patron,z1_f)
plt.grid()
plt.show()
```

Hago una FFT

La función fft, que devuelve la DFT de la señal de entrada. 
La función fftfreq, que devuelve las frecuencias a las que corresponde cada punto de la DFT. 
Se usa conjuntamente con la función anterior

```python
z1_fft=np.fft.fft(z1_f)
```

```python
freq = np.fft.fftfreq(Tiempo.shape[-1])
```

```python
print(100*freq[0:20])
```

```python
plt.figure(figsize=(15,8))
plt.plot(100*freq[0:500],np.abs(z1_fft[0:500]))
plt.grid()
plt.show()
```

Esto muestra la existncia de un sistema no lineal ,,, el pico entre los 4 Hz y 6 Hz,,, 
indica un subsistema de alta frecuencia,,, que no lo estudiaremos 

```python
#defino la función de auto-correlación
def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result[result.size // 2:]
```

```python
z1_corr=autocorr(z1_f)
```

```python
plt.figure(figsize=(15,8))
plt.plot(Tiempo,z1_corr)
plt.grid()
plt.show()
```
